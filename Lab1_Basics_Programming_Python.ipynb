# NumPy Exercises 

Now that we've learned about NumPy let's test your knowledge. We'll start off with a few simple tasks, and then you'll be asked some more complicated questions.

#### Import NumPy as np

````python
import numpy as np
````

#### Create an array of 10 zeros 

````python
arrOfZeros = np.zeros(10)
print(f"Array of zeros: {arrOfZeros}")
````

#### Create an array of 10 ones

````python
arrOfOnes = np.ones(10)
print(f"Array of ones: {arrOfOnes}")
````

#### Create an array of 10 fives

````python
arr1 = np.full(10,5)
print(f"Array: {arr1}")
````

#### Create an array of the integers from 10 to 50

````python

arr2 = np.arange(10,51)
print(f"Array: {arr2}")
````

#### Create an array of all the even integers from 10 to 50

````python
arr3 = np.arange(10,51,2)
print(f"Array: {arr3}")
````

#### Create a 3x3 matrix with values ranging from 0 to 8

````python
matrix = np.arange(0,9)
matrix = matrix.reshape(3,3)
print(f"Matrix: \n{matrix}")
````

#### Create a 3x3 identity matrix

````python
idMatrix = np.identity(3)
print(f"Matrix: \n{matrix}")
````

#### Use NumPy to generate a random number between 0 and 1

````python
randomInt = np.random.random()
print(f"Random Number: \n{randomInt}")
````

#### Use NumPy to generate an array of 25 random numbers sampled from a standard normal distribution

````python
randomArray = np.random.normal(loc=5, scale=0.5, size=25)

#randomArray = np.random.normal(loc=5, scale=0.5, size=(5,5)) For a 5x5 matrix

randomArray
````

#### Create an array of 20 linearly spaced points between 0 and 1:

````python
linSpacedArray = np.linspace(0,1,20)
print(f"Linearly Spaced Between 0 and 1:\n {linSpacedArray}")
````

### Numpy Indexing and Selection

Now you will be given a few matrices, and be asked to replicate the resulting matrix outputs:

````python
import numpy as np

arr_2d = np.array([
    [5, 10, 15],
    [20, 25, 30],
    [35, 40, 45]
])

arr_2d
````

Retrieve the value 25 from the array using indexing.

````python
val = arr_2d[1][1]
print(val)
````

Extract the second row of the array.

````python
row = arr_2d[1][:]
row
````

Extract the third column of the array.

````python
row = arr_2d.transpose()[2][:]
row
````

Extract the bottom-right 2Ã—2 sub-array from arr_2d.

````python
subArray = arr_2d[-2:, 1:]
subArray
````


# Task #2: Loading and Visualizing Images


a.	Loading/reading image using OpenCv

````python
import cv2
img = cv2.imread('data/Lab1_Image_Bunny.jpg') #Load the image data into a NumPy array

import matplotlib.pyplot as plt

#cv2: BGR -> plt: RGB 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGBA) #matplotlib uses RGB, and cv2 uses BGR, so the cvtColor function is used to convert the image channels
plt.imshow(img)
plt.axis("Off") # To remove the numbers/plot 
plt.show()
````

b.	Loading/reading image using PIL

````python
from PIL import Image
img2 = Image.open('data/Lab1_Image_Bunny.jpg') #load image into a PIL Image object
import matplotlib.pyplot as plt
import matplotlib.cm as cm
plt.imshow(img2)
plt.axis('Off')
plt.title("Bunny")
plt.show()
````
