```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("cat.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #BGR -> RGB
if img is None:
    print("No image found")

plt.axis('Off') #Removes the axis numbers
plt.imshow(img)
plt.show
```

# ***Task 1***

#### Part 1

```python
img_Resize = cv2.resize(img,dsize=(200,150))
plt.imshow(img_Resize)
plt.show
print(img_Resize.shape)
```

#### Part 2

```python
import numpy as np
rows , cols, channels = img.shape
angle = np.radians(120)
M = cv2.getRotationMatrix2D((rows//2, cols//2), 120,1)
img_Rotation = cv2.warpAffine(img, M, (cols,rows))
plt.axis('Off')
plt.imshow(img_Rotation)
plt.show()
```

#### Part 3

```python
shear_M = np.float32([[1, 0.5, 0],
             		[0, 1  , 0],
            		[0, 0  , 1]])
sheared_img = cv2.warpPerspective(img,shear_M,(int(cols*1.5),int(rows*1.5)))


plt.axis('Off')
plt.imshow(sheared_img)
plt.show()
```
